```python
from olympus.utils.data_transformer import DataTransformer
import numpy as np
import seaborn as sns
```

```python
data = np.array([np.random.uniform(low=2.0, high=3.0, size=100),
                 np.random.uniform(low=5.0, high=10.0, size=100)]).T

sns.distplot([x[0] for x in data])
sns.distplot([x[1] for x in data])
```

```python
t = DataTransformer(transformation='standardization')
t.train(data)
```

```python
data_transformed = t.transform(data)
sns.distplot([x[0] for x in data_transformed])
sns.distplot([x[1] for x in data_transformed])
```

```python
data2 = np.array([np.random.uniform(low=2.0, high=3.0, size=10),
                  np.random.uniform(low=5.0, high=10.0, size=10)]).T
sns.distplot([x[0] for x in data2])
sns.distplot([x[1] for x in data2])
```
